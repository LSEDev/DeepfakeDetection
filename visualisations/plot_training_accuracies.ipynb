# Training and validation accuracies

```python
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

af_dir = '../../all_faces_bucket/'
```

```python
def plot_accuracies(models):
    '''Plots training and validation accuracies.
    
    Input:
    models: Takes a list of models to be included in the plot.'''

    all_acc = {}
    
    # Create a dictionary of all models and their corresponding accuracies
    for each_model in models:
        all_acc[each_model] = pickle.load(open(af_dir + 'trained_models/training_accuracies/' + each_model, "rb"))
        
    # to fix
    epochs=25
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(24,8))  
                                   
    for each_model in all_acc:
        ax1.plot(all_acc[each_model]['accuracy'],label=each_model, color='blue')
          
        
    for each_model in all_acc:
        ax2.plot(all_acc[each_model]['val_accuracy'],label=each_model, color='blue')
        
    ax1.set_title("Training accuracy vs epochs")
    ax1.set_xlabel('Epochs')
    ax1.set_ylabel('Accuracy')
    ax1.legend()

    ax2.set_title("Validation accuracy vs epochs")
    ax2.set_xlabel('Epochs')
    ax2.set_ylabel('Accuracy')
    ax2.legend()
    

        
    #     ax1.set_ylim(ymin=1.25e+13)  
    #     ax1.set_ylim(ymax=3.2e+13)  
    #     ax1.xaxis.set_major_locator(plt.MaxNLocator(9))
    #     ax1.axvline(last_train_date, color='g', linestyle='dashed', linewidth=1.5)
    #     ax1.ticklabel_format(style='sci', axis='y', scilimits=(12,12))
    #     plt.sca(ax1)
    #     plt.xticks(rotation=10)

#     return all_acc
plot_accuracies(['vgg'])
```
