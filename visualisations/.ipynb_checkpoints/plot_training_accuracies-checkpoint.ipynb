# Training and validation accuracies

```python
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

af_dir = '../../all_faces_bucket/'
```

```python
def plot_accuracies(models):
    '''Plots training and validation accuracies.
    
    Input:
    models: Takes a list of models to be included in the plot.'''

    train_acc = {}
    val_acc = {}
    train_loss = {}
    val_loss = {}
    colours = ['blue','green','red','brown','orange', 'black']
    
    # Create a dictionary of all models and their corresponding accuracies
    for each_model in models:
        train_acc[each_model] = list(pd.read_csv(af_dir + 'trained_models/training_accuracies/' + each_model + '.csv')['accuracy'])
        val_acc[each_model] = list(pd.read_csv(af_dir + 'trained_models/training_accuracies/' + each_model + '.csv')['val_accuracy'])
        
    # to fix
    epochs=25
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(24,8))  
                                   
    for i, each_model in enumerate(train_acc):
        ax1.plot(train_acc[each_model],label=each_model, color=colours[i])
          
        
    for i, each_model in enumerate(val_acc):
        ax2.plot(val_acc[each_model],label=each_model, color=colours[i])
        
    ax1.set_title("Training accuracy vs epochs")
    ax1.set_xlabel('Epochs')
    ax1.set_ylabel('Accuracy')
    ax1.legend()

    ax2.set_title("Validation accuracy vs epochs")
    ax2.set_xlabel('Epochs')
    ax2.set_ylabel('Accuracy')
    ax2.legend()
    


    
plot_accuracies(['vgg', 'xception', 'resnet50', 'mobilenet', 'efficientnet'])

# Note: Mobilenet's first 30 epochs didnt save so this is the graph from 30 epochs onwards
```
